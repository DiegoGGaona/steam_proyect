```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("steam.csv")
```

```python
df.columns
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["appid","median_playtime","average_playtime"], inplace=True)
```

```python
df["release_date"]= pd.to_datetime(df["release_date"])
```

```python
df["release_year"]= df["release_date"].dt.year
```

```python
df["price_dollars"]=df["price"]/100
```

```python
df[df["price_dollars"]<0]
```

```python
df=df[df["price_dollars"]>=0]
```

```python
df["positive_ratio"]=(df["positive_ratings"]/ (df["positive_ratings"]+df["negative_ratings"]))*100
```

```python
df.sort_values(by="price_dollars", ascending=False)
```

```python
df.drop(columns=["poaitive_ratio",], inplace=True)
```

```python
df.to_csv("steam_clean.csv", index=False)
```

//which genres dominate in steam?// 

```python
top_genres = df["genres"].value_counts().head(10)
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x=top_genres.values, y= top_genres.index)
plt.title("Top Genres on Steam")
plt.show()
```

```python
sns.scatterplot(data=df, x="price_dollars",y="positive_ratio")
plt.show()
```
